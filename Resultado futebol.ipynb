```python
import numpy as np
import random
import pandas as pd
import time
import matplotlib.pyplot as plt
```

```python
def prob_de_dois_times(ind_A, ind_B):
    ind_A_norm = ind_A + 2.5
    ind_B_norm = ind_B + 2.5
    prob_vitoria_A = ind_A_norm / (ind_A_norm + ind_B_norm)
    prob_vitoria_B = ind_B_norm / (ind_A_norm + ind_B_norm)
    lista_probs = [prob_vitoria_A, prob_vitoria_B]
    desv_prob = np.std(lista_probs)
    prob_empate = 0.5 - desv_prob
    prob_n_emp = 1 - prob_empate
    prob_A_n_emp = prob_vitoria_A*prob_n_emp
    prob_B_n_emp = prob_vitoria_B*prob_n_emp
    
    soma = prob_A_n_emp + prob_B_n_emp + prob_empate
    
    return prob_A_n_emp, prob_B_n_emp, prob_empate, soma
    
```

```python
def distribuicao_de_campeoes(nome_A, ind_A, nome_B, ind_B, nome_C, ind_C, nome_D, ind_D, n_campeonatos):
    lista_campeoes = []
    
    for i in range(n_campeonatos):
        campeao = fase_de_grupos_rapido(nome_A, ind_A, nome_B, ind_B, nome_C, ind_C, nome_D, ind_D)
        lista_campeoes.append(campeao)
    
    count, bins, ignored = plt.hist(lista_campeoes)
    plt.title('Distribuição dos campeoes')
    plt.ylabel('Número de Títulos')
    plt.savefig('Distribuição dos campeoes.png', dpi=100, format='png', facecolor='white')
    plt.show()
    
    titulos = pd.Series(lista_campeoes).value_counts()
    
    return titulos
```
