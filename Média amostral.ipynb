```python
import numpy as np
import random as rd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
testes = [10, 100, 1000, 10000, 100000, 1000000, 10000000]
```

```python
razoes = []

lista_1 = list(range(10000000))

rd.shuffle(lista_1)

media_total = np.mean(lista_1)

for j in testes:
    lista_2 = []
    
    for i in range(j):
        lista_2.append(lista_1[i])
    
    media_amostral = np.mean(lista_2)
    
    razao = media_amostral/media_total
    razoes.append(razao)
```

```python
testes_razoes = {"Testes": testes,
                "Razões": razoes
               }

df_testes_razoes = pd.DataFrame.from_dict(testes_razoes)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
df_testes_razoes.plot('Testes','Razões',legend=False, ax=ax1, kind = 'scatter', linewidth = 2.5, color = 'blue')
df_testes_razoes.plot('Testes','Razões',legend=False, ax=ax1, kind = 'line', linewidth = 2.5, color = 'black')
plt.xlabel('Tamanho do conjunto')
plt.ylabel('Razão entre a média amostral e a média total')
plt.xscale("log")
plt.savefig('Média amostral.png', dpi=100, format='png', facecolor='white')
plt.show()
```

```python
lista_3 = list(range(100000))

medias_1000 = []

testes_2 = []

for t in range(2000):
    lista_4 = []
    
    rd.shuffle(lista_3)
    
    for i in range(100):
        lista_4.append(lista_3[i])
    
    media_amostral = np.mean(lista_4)
    
    medias_1000.append(media_amostral)
    
    testes_2.append(t)
```

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(medias_1000, 30)
plt.title('Distribuição das médias amostrais')
plt.xlabel('Média da amostra')
plt.savefig('Distribuição das médias amostrais.png', dpi=100, format='png', facecolor='white')
plt.show()
```
